```python
#imports
import pandas as pd 
import numpy as np
```

```python
#load the dataframe
df = pd.read_excel("CarDataset.xlsx") 

#remove all duplicate records
df.drop_duplicates(keep=False, inplace=True)#maybe move this down 
```

```python
#further clean the data to the desired specifications
#need to convert the mileage and year to numbers

df['mileage'] = df['mileage'].astype('string') 
df['mileage'] = df['mileage'].str.replace('km','').str.replace(',','')
df['mileage'] = pd.to_numeric(df['mileage'])

df['trim'] = df['trim'].astype('string') 
df['trim'] = df['trim'].str.lower()

df['make'] = df['make'].astype('string') 
df['make'] = df['make'].str.lower()

df['model'] = df['model'].astype('string') 
df['model'] = df['model'].str.lower()
```

```python
#read in all the car trims that were scraped
df2 = pd.read_excel("CarTrimData.xlsx") 
```

```python
#view how the data is arranged
df2.head()
```

```python
#create function where we give "trim" as seen above and clean list of trims and find trim for each model 
def trim_finder(trims,clean_trims):
    
    row_list = [x for x in clean_trims if str(x) != 'nan']
    row_list = [element.lower() for element in row_list]
    row_list = row_list[2:]
    row_list.sort(key = len)
    row_list.reverse()
    
    #clean the trim data isolate the description of a car only to its trim
    final_ws = [(" " + element + " ") for element in row_list] 
    final_ct = [(" " + element + " ") for element in trims]
    final_ct = [w.replace('*', ' ') for w in final_ct ]
    final_ct = [w.replace('-', ' ') for w in final_ct ]
    final_ct = [w.replace('–', ' ') for w in final_ct ]
    final_ct = [w.replace(',', ' ') for w in final_ct ]
    final_ct = [w.replace('\\', " ") for w in final_ct ]
    final_ct = [w.replace('\t', " ") for w in final_ct ]
    final_ct = [w.replace('/', " ") for w in final_ct ]
    final_ct = [w.replace('|', " ") for w in final_ct ]
    final_ct = [w.replace('.', ' ') for w in final_ct ]
    final_ct = [w.replace('*', ' ') for w in final_ct ]
    final_ct = [w.replace('!', ' ') for w in final_ct ]
    final_ct = [w.replace('(', ' ') for w in final_ct ]
    final_ct = [w.replace(')', ' ') for w in final_ct ]
    final_ct = [w.replace('+', ' ') for w in final_ct ]
    final_ct = [w.replace('$', ' ') for w in final_ct ]
    
    #give each trim desription a precise trim name by comparing it to trim database
    final_trim = []
    count = 0
    for i in final_ct:
        count = 0
        for j in final_ws:       
            if j in i:
                final_trim.append(j)
                count = 1
                break
        if count == 0:
            final_trim.append('N/A')
            count = 0
            
    return final_trim
```

```python
def final_dictionary():
    
    Dict = {'toyota':['corolla','highlander','camry','prius','rav4'], 
            'tesla':['model 3','model s','model x', 'model y'],
             'porsche':['taycan'],
             'honda': ['accord','civic', 'odyssey','civic coupe', 'civic type r'],
             'acura': ['ilx', 'integra','mdx','nsx','rdx','tl','tlx'],
             'audi':['a3', 'a4','a5','a6','a7','a8','e-tron','q3','q4','q5','q7','q8','s3','s4','s5','s6','s7','s8','rs 3','rs 4','rs 5','rs 6','rs 7','tt','sq5','sq7','sq8'],
             'bmw':['3 series','4 series','5 series','6 series','7 series','8 series','i3','i4','i5','i7','i8','ix','m2','m3','m4','m6','m8','x3','x4','x5','x6','x7','z3','x3 m','x4 m','x5 m','x6 m','z3'],
             'cadillac':['escalade','xt4','xt5','xt6'],
             'chevrolet':['bolt ev','camaro','corvette','cruze'],
             'dodge':['challenger','charger'],
             'ford':['f-150 lightning','mustang mach-e'],
             'infiniti':['q50','q60','qx50','qx60','qx80'],
             'lexus':['is','nx','rx'], 
             'mazda':['cx-5','mazda3'],
             'mercedes-benz':['c-class','cla','s-class'], #could add more mercedes later
             'subaru':['wrx','crosstrek'],
             'volkswagen':['e-golf','golf','golf gti','golf r','jetta','passat','taos','atlas'] 
             }
    
    return Dict
```

```python
#Dict = {'toyota':['corolla'], 'honda': ['accord']}

Dict = final_dictionary()
df_cleaned_trim = []

for key,value in Dict.items():
        for i in value:
            
            make = key
            model = i
      
            #create list of trim descriptions for a specific make and model
            car_features = df[(df['make']== make) & (df['model']== model)]
            all_trim = car_features['trim'].tolist()

            #create list of actual trim name for a specific make and model
            car_trims = df2[(df2['make']== make) & (df2['model']== model)]
            row_list = car_trims.values.flatten().tolist()
            
            #create a list of cleaned vehicle trims from vehicle descriptions
            clean_trim = trim_finder(all_trim,row_list)
            
            #place new cleaned descriptions back into a dataframe
            new_df = df[(df['make']== make) & (df['model']== model)]
            new_df.insert(3, "final_trim", clean_trim, True)
            
            #remove anomoloes for the year
            q_low = new_df["year"].quantile(0.01)
            q_hi  = new_df["year"].quantile(0.99)
            df_filtered = new_df[(new_df["year"] < q_hi) & (new_df["year"] > q_low)]
            
            #remove anomolies for the price
            q_low = df_filtered["price"].quantile(0.01)
            q_hi  = df_filtered["price"].quantile(0.99)
            df_filtered1 = df_filtered[(df_filtered["price"] < q_hi) & (df_filtered["price"] > q_low)]
            
            #remove anomolies for the mileage 
            q_low = df_filtered1["mileage"].quantile(0.01)
            q_hi  = df_filtered1["mileage"].quantile(0.99)
            df_filtered2 = df_filtered1[(df_filtered1["mileage"] < q_hi) & (df_filtered1["mileage"] > q_low)]
                
            #add final cleaned data to a list
            final_df = df_filtered2.to_dict('list')
            df_cleaned_trim.append(final_df)
        

lst = []
#take list of dataframes and turn into once dataframe    
for i in range(len(df_cleaned_trim)):
    df_t = pd.DataFrame(df_cleaned_trim[i])
    lst.append(df_t)

full_df = pd.concat(lst)
```

```python
#clean up remaining dataframe issues
cleaned_df = full_df
adj_df = cleaned_df.reset_index()
index_lst = adj_df[adj_df['final_trim'] == 'N/A'].index
final_dataset = adj_df.drop(adj_df.index[index_lst])
final_dataset = final_dataset.drop('index', axis=1)

#see how the final_trim values have been created by extracting the trim from the trim column
final_dataset
```

```python
#placing the file location
file_name = "FinalData.xlsx"

#saving the dataframe to the excel file
final_dataset.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
